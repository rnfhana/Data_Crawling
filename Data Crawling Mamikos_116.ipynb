```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
```

```python
# INSTALL BROWSER  SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

# USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
# IT IS SIMILAR TO REQUEST.GET
# SIMPLY TO GET THE WEBSITE THAT WE'RE TRYIN TO SCRAPE
browser.get("https://mamikos.com/cari/surabaya-kota-surabaya-jawa-timur-indonesia/all/bulanan/0-15000000?rent=2&sort=price,-&price=10000-20000000&singgahsini=0")
```

```python
# MAKE IT BEAUTIFUL HTML
soup = BeautifulSoup(browser.page_source, "lxml")
```

```python
rows = soup.find_all('div', class_='kost-rc__content')
```

```python
data = []
for row in rows:
    label_html = row.find_all(class_='rc-overview__label')
    if len(label_html)>1:
        label = label_html[1].text.replace('\n', '').replace('\t', '')
    else:
        label = label_html[0].text.replace('\n', '').replace('\t', '')
    availability_element = row.find(class_='rc-overview__availability')
    availability = availability_element.text.replace('\n', '').replace('\t', '') if availability_element is not None else ""
    name = row.find(class_='rc-info__name').text.replace('\n', '').replace('\t', '')
    location = row.find(class_='rc-info__location').text.replace('\n', '').replace('\t', '')
    facilities = row.find(class_='rc-facilities').text.replace('\n', '').replace('\t', '')
    discount_element = row.find(class_='rc-price__additional-discount')
    discount = discount_element.text.replace('\n', '').replace('\t', '') if discount_element is not None else ""
    discount_price_element = row.find(class_='rc-price__additional-discount-price')
    discount_price = discount_price_element.text.replace('\n', '').replace('\t', '') if discount_price_element is not None else ""
    price = row.find(class_='rc-price__text').text.replace('\n', '').replace('\t', '')
    
    dt = {
        'Tipe Kost': label,
        'Kamar Tersedia': availability,
        'Nama Kost': name,
        'Lokasi': location,
        'Fasilitas': facilities,
        'Harga Asli': discount_price,
        'Diskon': discount,
        'Harga /bulan': price
    }
    data.append(dt)
```

```python
df = pd.DataFrame(data)
```
